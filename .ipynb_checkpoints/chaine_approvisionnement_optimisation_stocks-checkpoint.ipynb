# 🏭 Chaîne d'Approvisionnement en Fabrication — Optimisation des Stocks
### Problème métier : Réduire les ruptures de stock, les surcoûts de surstockage et améliorer l'efficacité des réapprovisionnements

---

**Questions auxquelles ce notebook répond :**
1. Quels SKUs sont en rupture critique ou en surstock ?
2. Quels sont les points de réapprovisionnement optimaux et les Quantités Économiques de Commande (QEC / EOQ) ?
3. Quels fournisseurs causent des perturbations dans les stocks ?
4. Quelles actions concrètes permettront de réduire les coûts de stockage et d'éviter les ruptures ?

---

## 📦 Section 1 — Configuration & Génération des Données Synthétiques

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)

# ── PARAMÈTRES ───────────────────────────────────────────
N_SKUS            = 40
N_JOURS           = 365
N_FOURNISSEURS    = 8
TAUX_STOCKAGE     = 0.25   # 25% du coût unitaire par an
COUT_COMMANDE     = 150    # Coût fixe par commande (€)

CATEGORIES    = ['Matières Premières', 'Composants', 'Sous-Ensembles', 'Emballage', 'MRO']
FOURNISSEURS  = [f'Fournisseur_{chr(65+i)}' for i in range(N_FOURNISSEURS)]

print('✅ Bibliothèques chargées.')
print(f'📊 Simulation de {N_SKUS} SKUs × {N_JOURS} jours sur {N_FOURNISSEURS} fournisseurs')
```

```python
# ── TABLE MAÎTRE DES SKUs ─────────────────────────────────
skus = pd.DataFrame({
    'SKU_ID'            : [f'SKU-{1000+i}' for i in range(N_SKUS)],
    'Categorie'         : np.random.choice(CATEGORIES, N_SKUS),
    'Fournisseur'       : np.random.choice(FOURNISSEURS, N_SKUS),
    'Cout_Unitaire'     : np.round(np.random.uniform(5, 500, N_SKUS), 2),
    'Delai_Livraison'   : np.random.randint(3, 30, N_SKUS),
    'Stock_Securite'    : np.random.randint(20, 200, N_SKUS),
    'Stock_Actuel'      : np.random.randint(0, 600, N_SKUS),
    'Capacite_Max'      : np.random.randint(500, 2000, N_SKUS),
})

# Types de demande : stable, saisonnière ou volatile
types_demande = np.random.choice(['stable', 'saisonniere', 'volatile'], N_SKUS, p=[0.4, 0.35, 0.25])
skus['Type_Demande']        = types_demande
skus['Demande_Moy_Jour']    = np.round(np.random.uniform(5, 80, N_SKUS), 1)

print(skus.head(8).to_string(index=False))
```

```python
# ── TRANSACTIONS JOURNALIÈRES (365 jours) ─────────────────
dates   = pd.date_range('2024-01-01', periods=N_JOURS, freq='D')
records = []

for _, sku in skus.iterrows():
    avg_d = sku['Demande_Moy_Jour']
    for t, date in enumerate(dates):
        if sku['Type_Demande'] == 'saisonniere':
            saison = 1 + 0.5 * np.sin(2 * np.pi * t / 365)
            demande = max(0, np.random.poisson(avg_d * saison))
        elif sku['Type_Demande'] == 'volatile':
            demande = max(0, int(np.random.normal(avg_d, avg_d * 0.6)))
        else:
            demande = max(0, np.random.poisson(avg_d))

        records.append({
            'Date'   : date,
            'SKU_ID' : sku['SKU_ID'],
            'Demande': demande,
        })

txn = pd.DataFrame(records)

# Agrégation hebdomadaire
txn['Semaine']      = txn['Date'].dt.to_period('W')
hebdo               = txn.groupby(['SKU_ID', 'Semaine'])['Demande'].sum().reset_index()
hebdo['Debut_Sem']  = hebdo['Semaine'].dt.start_time

print(f'✅ {len(txn):,} lignes de transactions journalières générées')
txn.head()
```

## 📊 Section 2 — Vue d'Ensemble de la Santé des Stocks

```python
# ── CALCUL DES KPIs DE STOCK ──────────────────────────────
demande_annuelle = txn.groupby('SKU_ID')['Demande'].sum().reset_index()
demande_annuelle.columns = ['SKU_ID', 'Demande_Annuelle']

skus = skus.merge(demande_annuelle, on='SKU_ID')

# Jours de couverture restants
skus['Jours_Couverture'] = np.where(
    skus['Demande_Moy_Jour'] > 0,
    skus['Stock_Actuel'] / skus['Demande_Moy_Jour'],
    999
)

# Classification du statut de stock
def classifier_stock(row):
    jc = row['Jours_Couverture']
    dl = row['Delai_Livraison']
    if row['Stock_Actuel'] == 0:
        return 'RUPTURE'
    elif jc < dl:
        return 'CRITIQUE'
    elif jc < dl * 1.5:
        return 'BAS'
    elif row['Stock_Actuel'] > row['Capacite_Max'] * 0.85:
        return 'SURSTOCK'
    else:
        return 'SAIN'

skus['Statut_Stock'] = skus.apply(classifier_stock, axis=1)

recap_statut = skus['Statut_Stock'].value_counts().reset_index()
print('\n📋 Résumé des statuts de stock :')
print(recap_statut.to_string(index=False))
```

```python
# ── GRAPHIQUE 1 : Distribution des Statuts de Stock ───────
palette = {
    'RUPTURE' : '#EF4444',
    'CRITIQUE': '#F97316',
    'BAS'     : '#EAB308',
    'SAIN'    : '#22C55E',
    'SURSTOCK': '#3B82F6',
}

fig = px.pie(
    recap_statut,
    names='Statut_Stock',
    values='count',
    title='📦 Santé des Stocks — Tous les SKUs',
    color='Statut_Stock',
    color_discrete_map=palette,
    hole=0.45
)
fig.update_layout(template='plotly_dark', height=420)
fig.show()
```

```python
# ── GRAPHIQUE 2 : Niveau de Stock vs Jours de Couverture ──
fig = px.scatter(
    skus,
    x='Stock_Actuel',
    y='Jours_Couverture',
    color='Statut_Stock',
    size='Cout_Unitaire',
    hover_data=['SKU_ID', 'Categorie', 'Fournisseur', 'Delai_Livraison'],
    color_discrete_map=palette,
    title='🔍 Stock Actuel vs Jours de Couverture (bulle = coût unitaire)',
    labels={'Stock_Actuel': 'Stock Actuel (unités)', 'Jours_Couverture': 'Jours de Couverture Restants'}
)
fig.add_hline(y=14, line_dash='dash', line_color='orange', annotation_text='⚠️ Seuil 2 semaines')
fig.add_hline(y=7,  line_dash='dash', line_color='red',    annotation_text='🚨 Seuil critique 1 semaine')
fig.update_layout(template='plotly_dark', height=480)
fig.show()
```

## 📐 Section 3 — Analyse de la Quantité Économique de Commande (QEC / EOQ)

```python
# ── FORMULE QEC : sqrt(2 × D × S / H) ────────────────────
# D = Demande annuelle, S = Coût de commande, H = Coût de stockage unitaire annuel

skus['Cout_Stockage_Unitaire'] = skus['Cout_Unitaire'] * TAUX_STOCKAGE

skus['QEC'] = np.sqrt(
    (2 * skus['Demande_Annuelle'] * COUT_COMMANDE) / skus['Cout_Stockage_Unitaire']
).round(0).astype(int)

# Point de Réapprovisionnement = (Demande moy. / jour × Délai) + Stock de Sécurité
skus['Point_Reappro'] = (
    skus['Demande_Moy_Jour'] * skus['Delai_Livraison'] + skus['Stock_Securite']
).round(0).astype(int)

# Indicateur de réapprovisionnement immédiat
skus['Reappro_Urgent'] = skus['Stock_Actuel'] <= skus['Point_Reappro']

# Coût de stockage annuel au niveau actuel
skus['Cout_Stockage_Annuel'] = (skus['Stock_Actuel'] * skus['Cout_Stockage_Unitaire']).round(2)

print(f"\n🚨 SKUs nécessitant un réapprovisionnement IMMÉDIAT : {skus['Reappro_Urgent'].sum()}")
print(f"💰 Coût de stockage annuel total (niveau actuel) : {skus['Cout_Stockage_Annuel'].sum():,.0f} €")

skus[['SKU_ID', 'Categorie', 'QEC', 'Point_Reappro', 'Stock_Actuel', 'Reappro_Urgent']].head(10)
```

```python
# ── GRAPHIQUE 3 : QEC vs Stock Actuel ─────────────────────
fig = go.Figure()

fig.add_trace(go.Bar(
    name='Stock Actuel',
    x=skus['SKU_ID'],
    y=skus['Stock_Actuel'],
    marker_color='#60A5FA',
    opacity=0.8
))
fig.add_trace(go.Bar(
    name='QEC (Quantité Optimale)',
    x=skus['SKU_ID'],
    y=skus['QEC'],
    marker_color='#34D399',
    opacity=0.8
))
fig.add_trace(go.Scatter(
    name='Point de Réappro.',
    x=skus['SKU_ID'],
    y=skus['Point_Reappro'],
    mode='markers',
    marker=dict(color='#F59E0B', size=7, symbol='diamond'),
))

fig.update_layout(
    title='📐 QEC vs Stock Actuel vs Point de Réapprovisionnement par SKU',
    barmode='group',
    template='plotly_dark',
    height=500,
    xaxis_tickangle=-45,
    xaxis_title='SKU',
    yaxis_title='Unités'
)
fig.show()
```

## 📈 Section 4 — Prévision de la Demande & Analyse des Tendances

```python
# ── TOP 6 SKUs à plus forte valeur (demande × coût) ───────
skus['Score_Valeur'] = skus['Demande_Annuelle'] * skus['Cout_Unitaire']
top6 = skus.nlargest(6, 'Score_Valeur')['SKU_ID'].tolist()

hebdo_top6 = hebdo[hebdo['SKU_ID'].isin(top6)]

fig = px.line(
    hebdo_top6,
    x='Debut_Sem',
    y='Demande',
    color='SKU_ID',
    title='📈 Tendances de Demande Hebdomadaire — Top 6 SKUs à Haute Valeur',
    labels={'Demande': 'Demande Hebdomadaire (unités)', 'Debut_Sem': 'Semaine'},
)
fig.update_layout(template='plotly_dark', height=440)
fig.show()
```

```python
# ── GRAPHIQUE 5 : Variabilité de la Demande (CV) ──────────
stats_demande = txn.groupby('SKU_ID')['Demande'].agg(['mean', 'std']).reset_index()
stats_demande.columns = ['SKU_ID', 'Demande_Moy', 'Ecart_Type']
stats_demande['CV'] = (stats_demande['Ecart_Type'] / stats_demande['Demande_Moy']).round(3)

skus = skus.merge(stats_demande[['SKU_ID', 'CV']], on='SKU_ID')

fig = px.scatter(
    skus,
    x='Demande_Moy_Jour',
    y='CV',
    color='Categorie',
    size='Cout_Stockage_Annuel',
    hover_data=['SKU_ID', 'Fournisseur', 'Statut_Stock'],
    title='📉 Variabilité de la Demande (CV) vs Demande Moy. Journalière',
    labels={'CV': 'Coefficient de Variation (volatilité)', 'Demande_Moy_Jour': 'Demande Moy. Journalière'}
)
fig.add_hline(y=0.5, line_dash='dot', line_color='orange', annotation_text='Seuil Haute Volatilité')
fig.update_layout(template='plotly_dark', height=460)
fig.show()
```

## 🏭 Section 5 — Analyse de la Performance des Fournisseurs

```python
# ── INDICATEURS FOURNISSEURS ──────────────────────────────
fournisseurs_df = skus.groupby('Fournisseur').agg(
    Nb_SKUs             = ('SKU_ID', 'count'),
    Delai_Moy           = ('Delai_Livraison', 'mean'),
    SKUs_Rupture        = ('Statut_Stock', lambda x: (x == 'RUPTURE').sum()),
    SKUs_Critiques      = ('Statut_Stock', lambda x: (x == 'CRITIQUE').sum()),
    Cout_Stockage_Total = ('Cout_Stockage_Annuel', 'sum'),
    Cout_Unitaire_Moy   = ('Cout_Unitaire', 'mean'),
).reset_index()

fournisseurs_df['Score_Risque'] = (
    fournisseurs_df['SKUs_Rupture']   * 3 +
    fournisseurs_df['SKUs_Critiques'] * 2 +
    fournisseurs_df['Delai_Moy']      * 0.5
).round(1)

print(fournisseurs_df.sort_values('Score_Risque', ascending=False).to_string(index=False))
```

```python
# ── GRAPHIQUE 6 : Risque Fournisseur vs Délai de Livraison ─
fig = px.scatter(
    fournisseurs_df,
    x='Delai_Moy',
    y='Score_Risque',
    size='Cout_Stockage_Total',
    color='Score_Risque',
    text='Fournisseur',
    title='🏭 Score de Risque Fournisseur vs Délai Moyen (bulle = coût de stockage)',
    color_continuous_scale='RdYlGn_r',
    labels={'Delai_Moy': 'Délai Moyen (jours)', 'Score_Risque': 'Score de Risque'}
)
fig.update_traces(textposition='top center')
fig.update_layout(template='plotly_dark', height=460)
fig.show()
```

```python
# ── GRAPHIQUE 7 : Coût de Stockage par Catégorie & Fournisseur ─
cat_fourn = skus.groupby(['Categorie', 'Fournisseur'])['Cout_Stockage_Annuel'].sum().reset_index()

fig = px.sunburst(
    cat_fourn,
    path=['Categorie', 'Fournisseur'],
    values='Cout_Stockage_Annuel',
    title='💰 Répartition du Coût de Stockage Annuel par Catégorie & Fournisseur',
    color='Cout_Stockage_Annuel',
    color_continuous_scale='Blues'
)
fig.update_layout(template='plotly_dark', height=520)
fig.show()
```

## 🔴 Section 6 — Analyse ABC (Classification des Stocks)

```python
# ── ANALYSE ABC ───────────────────────────────────────────
# A = 70% de la valeur cumulée, B = 20% suivants, C = 10% restants
skus_abc = skus.sort_values('Score_Valeur', ascending=False).copy()
skus_abc['Valeur_Cum']  = skus_abc['Score_Valeur'].cumsum()
valeur_totale           = skus_abc['Score_Valeur'].sum()
skus_abc['Pct_Cum']     = skus_abc['Valeur_Cum'] / valeur_totale * 100

def classe_abc(pct):
    if pct <= 70:  return 'A — Haute Valeur'
    elif pct <= 90: return 'B — Valeur Moyenne'
    else:           return 'C — Faible Valeur'

skus_abc['Classe_ABC'] = skus_abc['Pct_Cum'].apply(classe_abc)

resume_abc = skus_abc.groupby('Classe_ABC').agg(
    Nb_SKUs             = ('SKU_ID', 'count'),
    Valeur_Totale       = ('Score_Valeur', 'sum'),
    Cout_Stockage_Moy   = ('Cout_Stockage_Annuel', 'mean'),
).reset_index()

print(resume_abc.to_string(index=False))
```

```python
# ── GRAPHIQUE 8 : Courbe de Pareto ABC ────────────────────
fig = make_subplots(specs=[[{'secondary_y': True}]])

fig.add_trace(go.Bar(
    x=skus_abc['SKU_ID'],
    y=skus_abc['Score_Valeur'],
    name='Valeur Annuelle (€)',
    marker_color=skus_abc['Classe_ABC'].map({
        'A — Haute Valeur'   : '#EF4444',
        'B — Valeur Moyenne' : '#F97316',
        'C — Faible Valeur'  : '#22C55E'
    }),
), secondary_y=False)

fig.add_trace(go.Scatter(
    x=skus_abc['SKU_ID'],
    y=skus_abc['Pct_Cum'],
    name='% Cumulé',
    line=dict(color='white', width=2),
    mode='lines'
), secondary_y=True)

fig.add_hline(y=70, secondary_y=True, line_dash='dash', line_color='red',    annotation_text='Seuil A/B (70%)')
fig.add_hline(y=90, secondary_y=True, line_dash='dash', line_color='orange', annotation_text='Seuil B/C (90%)')

fig.update_layout(
    title='📊 Analyse de Pareto ABC — Valeur Annuelle par SKU',
    template='plotly_dark',
    height=500,
    xaxis_tickangle=-45,
    showlegend=True
)
fig.update_yaxes(title_text='Valeur Annuelle (€)', secondary_y=False)
fig.update_yaxes(title_text='% Cumulé', secondary_y=True)
fig.show()
```

## 🎯 Section 7 — Tableau de Bord des Recommandations

```python
# ── GÉNÉRATION DES RECOMMANDATIONS ───────────────────────
skus_complet = skus_abc.copy()

def generer_action(row):
    actions = []
    statut  = row['Statut_Stock']
    abc     = row['Classe_ABC']
    cv      = row['CV']

    if statut == 'RUPTURE':
        actions.append(f'🚨 COMMANDE URGENTE : Passer une commande de {row["QEC"]} unités immédiatement')
    elif statut == 'CRITIQUE':
        actions.append(f'⚠️ RÉAPPROVISIONNEMENT URGENT : Stock inférieur au tampon de délai — commander {row["QEC"]} unités')
    elif statut == 'BAS':
        actions.append(f'📦 RÉAPPROVISIONNER BIENTÔT : Point de réappro. atteint ({row["Point_Reappro"]} unités)')
    elif statut == 'SURSTOCK':
        excedent = int(row['Stock_Actuel'] - row['Capacite_Max'] * 0.7)
        cout     = round(excedent * row['Cout_Stockage_Unitaire'], 0)
        actions.append(f'💰 RÉDUIRE LE STOCK : {excedent} unités en excès, coûtant {cout} €/an en stockage')

    if cv > 0.5 and abc == 'A — Haute Valeur':
        actions.append('📈 HAUTE VOLATILITÉ : Augmenter le stock de sécurité ou passer en gestion partagée (VMI)')

    if row['Delai_Livraison'] > 20 and statut in ['CRITIQUE', 'BAS']:
        actions.append(f"🏭 LONG DÉLAI ({row["Delai_Livraison"]}j) : Envisager un double sourcing auprès d'un fournisseur de secours")

    return ' | '.join(actions) if actions else '✅ Aucune action requise'

skus_complet['Action_Recommandee'] = skus_complet.apply(generer_action, axis=1)

plan_priorite = skus_complet[skus_complet['Action_Recommandee'] != '✅ Aucune action requise'].sort_values(
    ['Statut_Stock', 'Score_Valeur'], ascending=[True, False]
)[['SKU_ID', 'Categorie', 'Fournisseur', 'Classe_ABC', 'Statut_Stock',
   'Stock_Actuel', 'Point_Reappro', 'QEC', 'Jours_Couverture', 'Action_Recommandee']]

print(f'🎯 SKUs nécessitant une action : {len(plan_priorite)}')
plan_priorite.head(12)
```

```python
# ── GRAPHIQUE 9 : Carte de Chaleur — Catégorie × Statut ───
heat_data = skus_complet.pivot_table(
    index='Categorie',
    columns='Statut_Stock',
    values='SKU_ID',
    aggfunc='count',
    fill_value=0
)

fig = px.imshow(
    heat_data,
    text_auto=True,
    color_continuous_scale='RdYlGn_r',
    title='🔥 Carte de Chaleur des Stocks — Catégorie × Statut (nombre de SKUs)',
    aspect='auto'
)
fig.update_layout(template='plotly_dark', height=420)
fig.show()
```

```python
# ── GRAPHIQUE 10 : Économies Potentielles sur le Surstock ─
surstock_skus = skus_complet[skus_complet['Statut_Stock'] == 'SURSTOCK'].copy()
surstock_skus['Unites_Exces'] = (
    surstock_skus['Stock_Actuel'] - surstock_skus['Capacite_Max'] * 0.7
).clip(lower=0).astype(int)
surstock_skus['Economies_Potentielles'] = (
    surstock_skus['Unites_Exces'] * surstock_skus['Cout_Stockage_Unitaire']
).round(2)

if len(surstock_skus) > 0:
    fig = px.bar(
        surstock_skus.sort_values('Economies_Potentielles', ascending=True),
        x='Economies_Potentielles',
        y='SKU_ID',
        orientation='h',
        color='Categorie',
        title=f'💰 Économies Annuelles Potentielles en Réduisant le Surstock (Total : {surstock_skus["Economies_Potentielles"].sum():,.0f} €)',
        labels={'Economies_Potentielles': 'Économie Annuelle (€)', 'SKU_ID': 'SKU'}
    )
    fig.update_layout(template='plotly_dark', height=420)
    fig.show()
else:
    print('Aucun SKU en surstock détecté dans cette simulation.')
```

## ✅ Section 8 — Synthèse Exécutive & Plan d'Action

```python
# ── SYNTHÈSE EXÉCUTIVE ────────────────────────────────────
ruptures      = (skus_complet['Statut_Stock'] == 'RUPTURE').sum()
critiques     = (skus_complet['Statut_Stock'] == 'CRITIQUE').sum()
surstocks     = (skus_complet['Statut_Stock'] == 'SURSTOCK').sum()
cout_total    = skus_complet['Cout_Stockage_Annuel'].sum()
economies     = surstock_skus['Economies_Potentielles'].sum() if len(surstock_skus) > 0 else 0
fourn_risque  = fournisseurs_df.sort_values('Score_Risque', ascending=False).iloc[0]['Fournisseur']
skus_urgents  = plan_priorite[plan_priorite['Statut_Stock'].isin(['RUPTURE','CRITIQUE'])]['SKU_ID'].tolist()

synthese = f"""
╔══════════════════════════════════════════════════════════════════╗
║      🏭 OPTIMISATION DES STOCKS — SYNTHÈSE EXÉCUTIVE            ║
╠══════════════════════════════════════════════════════════════════╣
║  📊 ÉTAT ACTUEL                                                  ║
║  • Total SKUs analysés       : {N_SKUS}                                ║
║  • SKUs en rupture           : {ruptures} (action IMMÉDIATE requise)         ║
║  • SKUs en état critique     : {critiques} (commander dans le délai)         ║
║  • SKUs en surstock          : {surstocks}                                   ║
║  • Coût de stockage annuel   : {cout_total:>12,.0f} €                   ║
╠══════════════════════════════════════════════════════════════════╣
║  🎯 ACTIONS PRIORITAIRES                                         ║
║  1. 🚨 Commandes urgentes : {len(skus_urgents)} SKUs à réapprovisionner immédiatement ║
║  2. 💰 Réduction surstock : Économie potentielle de {economies:,.0f} €/an    ║
║  3. 🏭 Auditer {fourn_risque} : Score de risque le plus élevé          ║
║  4. 📐 Adopter le modèle QEC pour tous les SKUs de classe A      ║
║  5. 📈 Augmenter le stock de sécurité pour SKUs volatils (CV>0.5)║
║  6. 🔄 Mettre en place des alertes automatiques de réappro.      ║
╠══════════════════════════════════════════════════════════════════╣
║  📋 RECOMMANDATIONS STRATÉGIQUES                                 ║
║  • Revue ABC échelonnée : A hebdo, B bi-hebdo, C mensuel         ║
║  • Double sourcing pour fournisseurs avec délai > 20 jours       ║
║  • Prévision de demande renforcée pour SKUs volatils (CV > 0.5)  ║
║  • Négocier la gestion partagée (VMI) avec les 3 top fournisseurs║
╚══════════════════════════════════════════════════════════════════╝
"""
print(synthese)
```

```python
# ── EXPORT DU PLAN D'ACTION EN CSV ───────────────────────
plan_priorite.to_csv('plan_action_stocks.csv', index=False)
print('✅ Plan d\'action exporté vers : plan_action_stocks.csv')
print(f'   → {len(plan_priorite)} SKUs signalés avec des actions recommandées spécifiques')
```
