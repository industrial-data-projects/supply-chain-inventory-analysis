## 🧩 PART 1 — The Business Problem We Are Solving

---

### What is the problem?

Imagine a manufacturing company that produces, say, industrial machinery. To build those machines, it needs hundreds of parts: bolts, motors, sensors, packaging, lubricants, etc.

Every single one of those parts has to be **in stock at the right time, in the right quantity**.

If you have **too little** → the production line stops. Workers are idle. Customers wait. You lose money.

If you have **too much** → you paid for materials that are just sitting in a warehouse. That warehouse costs money to rent, heat, and manage. The materials may expire or become obsolete. You lose money.

**The goal of Inventory Optimization is to find the perfect middle ground:**
> *Always have enough stock to meet production demand — but never so much that you waste money storing it.*

---

### Why is this hard?

- Demand is **unpredictable** — some months you need 1,000 units, some months 3,000
- Suppliers have **lead times** — you order today but receive the goods in 2 weeks
- Each part has a **different cost**, a **different supplier**, a **different usage rate**
- You have **hundreds of SKUs** to manage simultaneously
- Cash is limited — you cannot simply buy enormous quantities of everything "just in case"

---

### What does the project answer?

| Business Question | How the project answers it |
|---|---|
| Which parts are about to run out? | Inventory Status classification |
| How much should I order each time? | EOQ formula |
| When exactly should I place an order? | Reorder Point calculation |
| Which parts deserve the most attention? | ABC Analysis |
| Which suppliers are causing problems? | Supplier Risk Score |
| How much money am I wasting on overstock? | Holding Cost analysis |

## 🏷️ PART 2 — Key Terms Dictionary (A to Z)
---

### 📌 SKU — Stock Keeping Unit

**Plain language:** A unique ID assigned to one specific product or material.

Think of it like a fingerprint for each item in your warehouse. Two bolts that look identical but come from different suppliers, or have different sizes, will have **different SKU codes** because they are treated as different items in the inventory system.

**Example:**
```
SKU-1001 → Steel Bolt M8, Supplier A
SKU-1002 → Steel Bolt M8, Supplier B  ← same bolt, different supplier = different SKU
SKU-1003 → Steel Bolt M10, Supplier A ← different size = different SKU
```

In our project we simulate **40 SKUs**, each representing a different material or component in a manufacturing facility.

---

### 📌 Demand

**Plain language:** How many units of a SKU are consumed (used up) per day by the production line.

In real life, demand is rarely constant. Our project simulates three realistic demand patterns:

| Demand Type | What it means | Real example |
|---|---|---|
| **Stable** | Consistent consumption every day | Standard fasteners, basic lubricants |
| **Seasonal** | High in some months, low in others | AC components, heating elements |
| **Volatile** | Unpredictable, big swings day to day | Special components for custom orders |

**Why it matters for cost control:** If demand spikes unexpectedly and you have no buffer, production stops — which is far more expensive than holding a bit of extra inventory.

---

### 📌 Lead Time

**Plain language:** The number of days between placing a purchase order and actually receiving the goods in your warehouse.

```
Day 0:  You place an order
Day 14: Goods arrive  ←  Lead Time = 14 days
```

**Why it matters:** You cannot wait until stock hits zero to order. You need to order **before** you run out, with enough time for the goods to arrive. Lead time tells you how early you must act.

In our project, SKUs have lead times ranging from **3 to 30 days**, simulating a mix of local and international suppliers.

---

### 📌 Safety Stock

**Plain language:** A small emergency buffer of inventory kept on hand specifically to absorb unexpected demand spikes or supplier delays.

It is the inventory equivalent of keeping a spare tire in your car — you hope to never need it, but it saves you when something goes wrong.

```
Without safety stock: demand spike → stockout → line stops
With safety stock:    demand spike → buffer absorbs it → line keeps running
```

**Cost controller perspective:** Safety stock has a cost (you are holding units you may not immediately need), but the cost of a stockout — idle labor, missed delivery penalties, emergency procurement — is almost always higher.

---

### 📌 Reorder Point (ROP)

**Plain language:** The stock level at which you must immediately place a new order — not because you have run out, but because if you wait any longer, you will run out before the new order arrives.

**Formula:**
```
Reorder Point = (Average Daily Demand × Lead Time in Days) + Safety Stock
```

**Example:**
```
Average Daily Demand = 50 units/day
Lead Time            = 10 days
Safety Stock         = 100 units

Reorder Point = (50 × 10) + 100 = 600 units

→ When stock drops to 600 units, place the order.
→ By the time the order arrives (10 days), you will have used ~500 units.
→ The remaining 100 units (safety stock) covers any demand surprises.
```

---

### 📌 EOQ — Economic Order Quantity

**Plain language:** The mathematically optimal quantity to order each time you place a purchase order — the quantity that minimizes your total inventory cost.

**The trade-off EOQ solves:**
- Order **too little, too often** → you pay more in ordering/shipping costs
- Order **too much, rarely** → you pay more in storage/holding costs
- EOQ finds the **sweet spot** between these two

**Formula:**
```
EOQ = √( 2 × D × S / H )

Where:
  D = Annual Demand (units/year)
  S = Order Cost (fixed cost per order, e.g. $150)
  H = Holding Cost per unit per year (unit cost × holding rate)
```

**Example:**
```
Annual Demand  = 18,000 units
Order Cost     = $150 per order
Unit Cost      = $20
Holding Rate   = 25% per year → Holding Cost = $5/unit/year

EOQ = √(2 × 18,000 × 150 / 5) = √(1,080,000) ≈ 1,039 units per order
```

**Cost controller perspective:** EOQ directly impacts your procurement budget. Ordering too frequently means higher PO processing, transport, and admin costs. Ordering too much means higher warehouse costs. EOQ is your lever to reduce both simultaneously.

---

### 📌 Holding Cost (Carrying Cost)

**Plain language:** The total cost of keeping one unit of inventory in your warehouse for one year.

It includes:
- Warehouse rent / space allocation
- Insurance on the stored goods
- Risk of obsolescence or expiry
- Opportunity cost (the money tied up in stock could be invested elsewhere)
- Handling and counting labor

**Typical rule of thumb:** Holding cost ≈ 20–30% of the item's unit cost per year.

In our project we use **25%**.

**Example:** A component costing $200 costs you $50/year just to store it. If you are holding 500 units unnecessarily, that is $25,000/year wasted.

---

### 📌 Stockout

**Plain language:** You have zero units of a SKU left in stock. The production line needs it but it is not there.

**Consequences in manufacturing:**
- Production line stoppage
- Idle workers (paying salaries for zero output)
- Emergency procurement at premium prices
- Delayed delivery to customers → penalties
- Damaged supplier/customer relationships

Stockout is the most expensive inventory failure. In our project, these SKUs are flagged as **🚨 EMERGENCY ORDER** in the action plan.

---

### 📌 Overstock

**Plain language:** You have far more units in stock than you need. The excess is just sitting in the warehouse costing you money.

**Causes:**
- Poor demand forecasting (ordered based on wrong assumptions)
- Bulk purchasing without considering storage cost
- Supplier minimum order quantities that are too high
- Demand suddenly dropped (product redesign, contract loss)

**Cost controller perspective:** Overstock is a hidden cost that shows up as warehouse cost and working capital tied up in materials. The project quantifies exactly how much overstock is costing each year per SKU.

---

### 📌 Days of Supply (DOS)

**Plain language:** Given the current stock level, how many days until you run out — assuming average daily demand continues.

**Formula:**
```
Days of Supply = Current Stock ÷ Average Daily Demand
```

**Example:**
```
Current Stock       = 300 units
Avg Daily Demand    = 25 units/day
Days of Supply      = 300 ÷ 25 = 12 days
```

If lead time is 15 days → you should have ordered 3 days ago. This SKU is **CRITICAL**.

---

### 📌 Inventory Status Classification

The project assigns each SKU one of five health statuses:

| Status | Meaning | Action |
|---|---|---|
| 🔴 **STOCKOUT** | Zero stock. Line at risk. | Emergency order NOW |
| 🟠 **CRITICAL** | Days of Supply < Lead Time | Order immediately |
| 🟡 **LOW** | Days of Supply < 1.5× Lead Time | Order soon |
| 🟢 **HEALTHY** | Sufficient stock, not excessive | Monitor normally |
| 🔵 **OVERSTOCK** | Stock > 85% of max warehouse capacity | Pause ordering, reduce |

---

### 📌 ABC Analysis

**Plain language:** Not all SKUs deserve equal attention. ABC Analysis ranks them by their total annual value (demand × unit cost) so you focus your energy where it matters most.

| Class | Rule | What to do |
|---|---|---|
| **A — High Value** | Top 70% of total inventory value | Review weekly, tight control, use EOQ |
| **B — Medium Value** | Next 20% | Review bi-weekly, standard monitoring |
| **C — Low Value** | Bottom 10% | Review monthly, simpler rules |

**Classic Pareto insight:** Typically, ~20% of your SKUs account for ~80% of your inventory value. Those are your A-class items.

**Cost controller perspective:** A-class SKUs are where errors in ordering are most expensive. A 10% overstock on a $500/unit item costs far more than a 10% overstock on a $5/unit item.

---

### 📌 CV — Coefficient of Variation

**Plain language:** A number that tells you how volatile (unpredictable) a SKU's demand is. The higher the CV, the harder it is to forecast and the more safety stock you need.

**Formula:**
```
CV = Standard Deviation of Demand ÷ Mean Demand
```

| CV Value | Interpretation |
|---|---|
| < 0.2 | Very stable demand — easy to manage |
| 0.2 – 0.5 | Moderate variability — manageable |
| > 0.5 | High volatility — needs larger safety stock or special strategy |

In our project, SKUs with CV > 0.5 AND Class A status get a recommendation to increase safety stock or switch to vendor-managed inventory.

---

### 📌 Supplier Risk Score

**Plain language:** A composite score we calculate to identify which suppliers are causing the most inventory problems.

**How it is calculated in the project:**
```
Risk Score = (Stockout SKUs × 3) + (Critical SKUs × 2) + (Avg Lead Time × 0.5)
```

- Stockouts are weighted most heavily (×3) because they cause production stoppages
- Critical SKUs are weighted ×2 (about to cause stoppages)
- Long lead times add baseline risk (×0.5)

**Why this matters for a cost controller:** Supplier problems translate directly to emergency procurement costs, expediting fees, and production losses. The risk score tells you which supplier relationships to audit first.

---

### 📌 VMI — Vendor Managed Inventory

**Plain language:** Instead of you managing when and how much to reorder from a supplier, the supplier takes responsibility for keeping your stock at agreed-upon levels. They monitor your consumption data and replenish automatically.

**Best for:** High-volume, stable-demand, A-class items with a trusted supplier.

**Benefit:** Reduces your ordering workload and often reduces stockouts because the supplier has better visibility into their own production schedule.

---

### 📌 MRO — Maintenance, Repair & Operations

**Plain language:** All the consumable supplies and parts used to keep the factory running that are NOT part of the final product.

**Examples:** Machine oil, gloves, cleaning chemicals, spare machine parts, light bulbs, tools.

MRO items are often C-class in ABC analysis but a stockout of a critical MRO item (e.g., a specific machine fuse) can still stop an entire production line. A good cost controller keeps an eye on critical MRO even if its total value is low.

---

### 📌 Sub-Assemblies

**Plain language:** Partially assembled components that will be combined later to make the final product. They sit between raw materials and finished goods in the production hierarchy.

**Example in manufacturing:**
```
Steel + Bolts → Welded Frame (Sub-Assembly) → Combined with Engine + Casing → Final Machine
```

Sub-assembly inventory optimization is more complex because stockouts can cascade up the production chain.

## 📈 PART 4 — Understanding Each Chart
---

### Chart 1 — Inventory Health Status (Donut Chart)
**What it shows:** What percentage of your SKUs fall into each health category right now.

**What to look for:** A healthy operation should have mostly GREEN (Healthy). Too much RED (Stockout/Critical) means procurement is reactive, not proactive. Too much BLUE (Overstock) means purchasing is not synchronized with actual demand.

---

### Chart 2 — Current Stock vs Days of Supply (Bubble Scatter)
**What it shows:** Every SKU as a bubble. X-axis = how much stock you have. Y-axis = how many days that stock will last. Bubble size = unit cost (bigger bubble = more expensive item).

**What to look for:** Expensive bubbles (large) that are BELOW the red 7-day line are your most dangerous situation — high-cost items that are about to stock out.

---

### Chart 3 — EOQ vs Current Stock vs Reorder Point (Grouped Bar)
**What it shows:** For each SKU, a comparison of three key numbers side by side.

**What to look for:**
- Blue bar (Current Stock) below the diamond (Reorder Point) → order NOW
- Blue bar much higher than green bar (EOQ) → you ordered too much last time

---

### Chart 4 — Weekly Demand Trends (Line Chart)
**What it shows:** How demand has evolved over 52 weeks for the 6 highest-value SKUs.

**What to look for:** Seasonal patterns (waves), sudden spikes, or trends going up/down. These patterns should inform how much safety stock you carry and when.

---

### Chart 5 — Demand Variability / CV (Scatter)
**What it shows:** Each SKU plotted by how much it is used daily vs how unpredictable it is. Color = category, size = holding cost.

**What to look for:** SKUs in the top-right area (high demand AND high variability) are your highest-risk items for stockouts. They need more safety stock.

---

### Chart 6 — Supplier Risk vs Lead Time (Bubble Scatter)
**What it shows:** Each supplier plotted by their average lead time and their computed risk score. Bubble size = how much holding cost their SKUs generate.

**What to look for:** Suppliers in the top-right (long lead time + high risk) are your most dangerous dependencies. Consider dual-sourcing or negotiating better terms.

---

### Chart 7 — Holding Cost Sunburst
**What it shows:** A nested circle diagram showing the breakdown of annual holding costs. Outer ring = supplier. Inner ring = category.

**What to look for:** Which category/supplier combination is costing you the most in storage? That is where a reduction in order quantities would save the most money.

---

### Chart 8 — ABC Pareto (Bar + Line)
**What it shows:** SKUs sorted by annual value (bars), with a cumulative percentage line on top.

**What to look for:** The point where the line crosses 70% marks the boundary between Class A and Class B items. Usually a small number of SKUs account for the majority of your inventory investment.

---

### Chart 9 — Inventory Status Heatmap
**What it shows:** A color grid. Rows = categories (Raw Materials, Components, etc.), Columns = inventory status. The number in each cell = how many SKUs are in that combination.

**What to look for:** Which category has the most stockouts or critical items? That tells you where procurement planning needs to improve.

---

### Chart 10 — Cost Savings from Overstock Reduction
**What it shows:** A horizontal bar chart of every overstocked SKU with the annual saving you would achieve by reducing it to 70% of warehouse capacity.

**What to look for:** The total at the top of the chart title tells you the total amount you could save per year just by not over-ordering.

## 🎯 PART 5 — How the Action Plan Works
---

The final section of the main notebook automatically generates a recommended action for every SKU that needs attention. Here is the decision logic:

```
FOR EACH SKU:

  IF Current Stock = 0
     → 🚨 EMERGENCY ORDER: Place order for [EOQ] units immediately

  ELSE IF Days of Supply < Lead Time
     → ⚠️ URGENT REORDER: Stock below lead-time buffer

  ELSE IF Days of Supply < 1.5 × Lead Time
     → 📦 REORDER SOON: Approaching reorder point

  ELSE IF Current Stock > 85% of Max Capacity
     → 💰 REDUCE STOCK: Calculate excess units and annual holding cost wasted

  ADDITIONALLY:
  IF CV > 0.5 AND ABC Class = A
     → 📈 HIGH VOLATILITY: Increase safety stock or use VMI

  IF Lead Time > 20 days AND Status is CRITICAL or LOW
     → 🏭 LONG LEAD TIME: Consider dual-sourcing from a backup supplier
```

This logic mirrors what an experienced supply chain planner would do manually — but it runs across all 40 SKUs in seconds.

## 🔑 PART 6 — Key Formulas Reference Card

| Formula | Equation |
|---|---|
| Days of Supply | Current Stock ÷ Avg Daily Demand |
| Reorder Point | (Avg Daily Demand × Lead Time) + Safety Stock |
| EOQ | √( 2 × Annual Demand × Order Cost ÷ Holding Cost per Unit ) |
| Holding Cost per Unit | Unit Cost × Holding Rate (e.g. 25%) |
| Annual Holding Cost | Current Stock × Holding Cost per Unit |
| Coefficient of Variation | Std Deviation of Demand ÷ Mean Demand |
| Supplier Risk Score | (Stockouts × 3) + (Critical SKUs × 2) + (Avg Lead Time × 0.5) |
| Annual Value (ABC) | Annual Demand × Unit Cost |
