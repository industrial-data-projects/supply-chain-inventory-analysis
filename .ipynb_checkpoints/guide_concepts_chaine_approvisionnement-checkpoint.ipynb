## 🧩 PARTIE 1 — Le Problème Métier que Nous Résolvons

---

### Quel est le problème ?

Imaginons une entreprise de fabrication industrielle qui produit, par exemple, des machines ou des équipements. Pour fabriquer ces produits, elle a besoin de centaines de pièces : boulons, moteurs, capteurs, emballages, lubrifiants, etc.

Chacune de ces pièces doit être **disponible au bon moment, en bonne quantité**.

Si vous en avez **trop peu** → la ligne de production s'arrête. Les opérateurs sont inactifs. Les clients attendent. Vous perdez de l'argent.

Si vous en avez **trop** → vous avez payé des matériaux qui restent dans un entrepôt. Cet entrepôt coûte en loyer, en énergie, en main-d'œuvre. Les matériaux peuvent devenir obsolètes ou se détériorer. Vous perdez de l'argent.

**L'objectif de l'Optimisation des Stocks est de trouver le juste équilibre :**
> *Toujours avoir assez de stock pour répondre à la demande de production — sans en avoir tellement que cela représente un gaspillage financier.*

---

### Pourquoi est-ce difficile ?

- La demande est **imprévisible** — certains mois vous avez besoin de 1 000 unités, d'autres mois de 3 000
- Les fournisseurs ont des **délais de livraison** — vous commandez aujourd'hui mais recevez les marchandises dans 2 semaines
- Chaque pièce a un **coût différent**, un **fournisseur différent**, un **rythme de consommation différent**
- Vous devez gérer **des centaines de références** simultanément
- La trésorerie est limitée — vous ne pouvez pas simplement acheter d'énormes quantités de tout "au cas où"

---

### À quelles questions le projet répond-il ?

| Question métier | Comment le projet y répond |
|---|---|
| Quelles pièces sont sur le point de manquer ? | Classification du statut de stock |
| Quelle quantité commander à chaque fois ? | Formule QEC (EOQ) |
| Quand exactement déclencher une commande ? | Calcul du Point de Réapprovisionnement |
| Quelles pièces méritent le plus d'attention ? | Analyse ABC |
| Quels fournisseurs causent des problèmes ? | Score de Risque Fournisseur |
| Combien d'argent gaspille-t-on en surstock ? | Analyse du Coût de Stockage |

## 🏷️ PARTIE 2 — Dictionnaire des Termes Clés (de A à Z)
---

### 📌 SKU — Stock Keeping Unit (Unité de Gestion de Stock)

**En langage clair :** Un identifiant unique attribué à un produit ou à un matériau spécifique.

C'est comme une empreinte digitale pour chaque article dans votre entrepôt. Deux boulons qui semblent identiques mais provenant de fournisseurs différents, ou de tailles différentes, auront des **codes SKU différents** car ils sont traités comme des articles distincts dans le système de gestion des stocks.

**Exemple :**
```
SKU-1001 → Boulon Acier M8, Fournisseur A
SKU-1002 → Boulon Acier M8, Fournisseur B  ← même boulon, fournisseur différent = SKU différent
SKU-1003 → Boulon Acier M10, Fournisseur A ← taille différente = SKU différent
```

Dans notre projet, nous simulons **40 SKUs**, chacun représentant un matériau ou un composant différent dans une usine.

---

### 📌 Demande (Demand)

**En langage clair :** Le nombre d'unités d'un SKU consommées (utilisées) par jour par la ligne de production.

En réalité, la demande est rarement constante. Notre projet simule trois schémas réalistes :

| Type de Demande | Ce que cela signifie | Exemple concret |
|---|---|---|
| **Stable** | Consommation régulière chaque jour | Visserie standard, lubrifiants de base |
| **Saisonnière** | Élevée certains mois, faible d'autres | Composants de climatisation, éléments chauffants |
| **Volatile** | Imprévisible, avec de grandes variations | Pièces spéciales pour commandes sur mesure |

**Pourquoi c'est important :** Si la demande augmente soudainement et que vous n'avez pas de tampon, la production s'arrête — ce qui coûte bien plus cher que de maintenir un petit stock supplémentaire.

---

### 📌 Délai de Livraison (Lead Time)

**En langage clair :** Le nombre de jours entre le passage d'une commande d'achat et la réception effective des marchandises dans votre entrepôt.

```
Jour  0 : Vous passez une commande
Jour 14 : Les marchandises arrivent  ←  Délai de livraison = 14 jours
```

**Pourquoi c'est important :** Vous ne pouvez pas attendre que le stock atteigne zéro pour commander. Vous devez commander **avant** d'être à court, avec suffisamment de temps pour que les marchandises arrivent. Le délai de livraison vous indique combien de temps à l'avance vous devez agir.

Dans notre projet, les SKUs ont des délais de **3 à 30 jours**, simulant un mix de fournisseurs locaux et internationaux.

---

### 📌 Stock de Sécurité (Safety Stock)

**En langage clair :** Un petit tampon d'inventaire d'urgence conservé spécifiquement pour absorber les pics de demande imprévus ou les retards de fournisseurs.

C'est l'équivalent en gestion de stocks d'avoir une roue de secours dans votre voiture — vous espérez ne jamais en avoir besoin, mais elle vous sauve quand quelque chose se passe mal.

```
Sans stock de sécurité : pic de demande → rupture → ligne arrêtée
Avec stock de sécurité : pic de demande → tampon absorbe → ligne continue
```

**Du point de vue de la gestion :** Le stock de sécurité a un coût (vous stockez des unités dont vous n'avez pas immédiatement besoin), mais le coût d'une rupture — main-d'œuvre inactive, pénalités de livraison manquée, approvisionnement d'urgence — est presque toujours bien plus élevé.

---

### 📌 Point de Réapprovisionnement (Reorder Point — ROP)

**En langage clair :** Le niveau de stock à partir duquel vous devez immédiatement passer une nouvelle commande — non pas parce que vous êtes à court, mais parce que si vous attendez plus longtemps, vous n'aurez plus de stock avant que la nouvelle commande n'arrive.

**Formule :**
```
Point de Réappro. = (Demande Moy. Journalière × Délai de Livraison en Jours) + Stock de Sécurité
```

**Exemple :**
```
Demande Moy. Journalière = 50 unités/jour
Délai de Livraison       = 10 jours
Stock de Sécurité        = 100 unités

Point de Réappro. = (50 × 10) + 100 = 600 unités

→ Quand le stock tombe à 600 unités, passez la commande.
→ Au moment où la commande arrivera (10 jours), vous aurez consommé ~500 unités.
→ Les 100 unités restantes (stock de sécurité) couvrent les imprévus de demande.
```

---

### 📌 QEC / EOQ — Quantité Économique de Commande (Economic Order Quantity)

**En langage clair :** La quantité mathématiquement optimale à commander à chaque fois que vous passez une commande — la quantité qui minimise votre coût total de gestion des stocks.

**Le compromis que la QEC résout :**
- Commander **trop peu, trop souvent** → vous payez plus en frais de commande et de transport
- Commander **trop, rarement** → vous payez plus en coûts de stockage
- La QEC trouve le **point optimal** entre ces deux extrêmes

**Formule :**
```
QEC = √( 2 × D × S / H )

Où :
  D = Demande Annuelle (unités/an)
  S = Coût de Commande (coût fixe par commande, ex. 150 €)
  H = Coût de Stockage par unité par an (coût unitaire × taux de stockage)
```

**Exemple :**
```
Demande Annuelle = 18 000 unités
Coût de Commande = 150 € par commande
Coût Unitaire    = 20 €
Taux de Stockage = 25% par an → Coût de Stockage = 5 €/unité/an

QEC = √(2 × 18 000 × 150 / 5) = √(1 080 000) ≈ 1 039 unités par commande
```

**Point de vue ingénierie/gestion :** La QEC impacte directement votre budget d'approvisionnement. Commander trop fréquemment signifie des frais de traitement de bons de commande, de transport et d'administration plus élevés. Commander trop grande quantité signifie des frais d'entrepôt plus élevés. La QEC est le levier pour réduire les deux simultanément.

---

### 📌 Coût de Stockage (Holding Cost / Carrying Cost)

**En langage clair :** Le coût total de conservation d'une unité d'inventaire dans votre entrepôt pendant un an.

Il comprend :
- Loyer de l'entrepôt / allocation d'espace
- Assurance sur les marchandises stockées
- Risque d'obsolescence ou d'expiration
- Coût d'opportunité (l'argent immobilisé en stock pourrait être investi ailleurs)
- Main-d'œuvre pour la manutention et les inventaires

**Règle empirique habituelle :** Le coût de stockage ≈ 20–30% du coût unitaire de l'article par an.

Dans notre projet, nous utilisons **25%**.

**Exemple :** Un composant coûtant 200 € vous coûte 50 €/an rien que pour le stocker. Si vous conservez 500 unités inutilement, cela représente 25 000 €/an gaspillés.

---

### 📌 Rupture de Stock (Stockout)

**En langage clair :** Vous avez zéro unité d'un SKU en stock. La ligne de production en a besoin mais ce n'est pas là.

**Conséquences en fabrication :**
- Arrêt de la ligne de production
- Opérateurs inactifs (salaires payés pour aucune production)
- Approvisionnement d'urgence à des prix majorés
- Livraison retardée aux clients → pénalités
- Relations dégradées avec les clients et les fournisseurs

La rupture de stock est l'échec de gestion des stocks le plus coûteux. Dans notre projet, ces SKUs sont signalés comme **🚨 COMMANDE URGENTE** dans le plan d'action.

---

### 📌 Surstock (Overstock)

**En langage clair :** Vous avez bien plus d'unités en stock que nécessaire. L'excédent reste dans l'entrepôt et vous coûte de l'argent.

**Causes fréquentes :**
- Mauvaise prévision de la demande (commande basée sur de mauvaises hypothèses)
- Achats en grande quantité sans considérer le coût de stockage
- Quantités minimales de commande imposées par les fournisseurs trop élevées
- Chute soudaine de la demande (reconception du produit, perte de contrat)

**Du point de vue de la gestion :** Le surstock est un coût caché qui se manifeste sous forme de coûts d'entrepôt et de capital immobilisé. Le projet quantifie exactement combien le surstock coûte chaque année par SKU.

---

### 📌 Jours de Couverture (Days of Supply — DOS)

**En langage clair :** Avec le niveau de stock actuel, dans combien de jours serez-vous à court — en supposant que la demande journalière moyenne continue.

**Formule :**
```
Jours de Couverture = Stock Actuel ÷ Demande Moy. Journalière
```

**Exemple :**
```
Stock Actuel          = 300 unités
Demande Moy. / Jour   = 25 unités/jour
Jours de Couverture   = 300 ÷ 25 = 12 jours
```

Si le délai de livraison est de 15 jours → vous auriez dû commander il y a 3 jours. Ce SKU est en état **CRITIQUE**.

---

### 📌 Classification du Statut de Stock

Le projet attribue à chaque SKU l'un des cinq statuts de santé suivants :

| Statut | Signification | Action requise |
|---|---|---|
| 🔴 **RUPTURE** | Stock à zéro. Ligne de production en danger. | Commande d'urgence IMMÉDIATEMENT |
| 🟠 **CRITIQUE** | Jours de Couverture < Délai de Livraison | Commander immédiatement |
| 🟡 **BAS** | Jours de Couverture < 1,5 × Délai de Livraison | Commander bientôt |
| 🟢 **SAIN** | Stock suffisant, non excessif | Surveiller normalement |
| 🔵 **SURSTOCK** | Stock > 85% de la capacité max de l'entrepôt | Suspendre les commandes, réduire |

---

### 📌 Analyse ABC

**En langage clair :** Tous les SKUs ne méritent pas la même attention. L'Analyse ABC les classe par leur valeur annuelle totale (demande × coût unitaire) pour concentrer les efforts là où c'est le plus important.

| Classe | Règle | Que faire |
|---|---|---|
| **A — Haute Valeur** | 70% premiers de la valeur totale des stocks | Révision hebdomadaire, contrôle serré, utiliser la QEC |
| **B — Valeur Moyenne** | 20% suivants | Révision bi-hebdomadaire, suivi standard |
| **C — Faible Valeur** | 10% restants | Révision mensuelle, règles simplifiées |

**Le principe de Pareto :** Typiquement, ~20% de vos SKUs représentent ~80% de la valeur de votre inventaire. Ce sont vos articles de classe A.

**En gestion industrielle :** Les SKUs de classe A sont là où les erreurs de commande sont les plus coûteuses. Un surstock de 10% sur un article à 500 € coûte bien plus qu'un surstock de 10% sur un article à 5 €.

---

### 📌 CV — Coefficient de Variation

**En langage clair :** Un nombre qui indique à quel point la demande d'un SKU est volatile (imprévisible). Plus le CV est élevé, plus il est difficile de faire une prévision et plus le stock de sécurité nécessaire est important.

**Formule :**
```
CV = Écart-Type de la Demande ÷ Demande Moyenne
```

| Valeur du CV | Interprétation |
|---|---|
| < 0,2 | Demande très stable — facile à gérer |
| 0,2 – 0,5 | Variabilité modérée — gérable |
| > 0,5 | Haute volatilité — nécessite un stock de sécurité plus important ou une stratégie spéciale |

Dans notre projet, les SKUs avec CV > 0,5 ET de classe A reçoivent une recommandation d'augmenter le stock de sécurité ou de passer en gestion partagée (VMI).

---

### 📌 Score de Risque Fournisseur

**En langage clair :** Un score composite calculé pour identifier quels fournisseurs causent le plus de problèmes de stock.

**Comment il est calculé dans le projet :**
```
Score de Risque = (SKUs en Rupture × 3) + (SKUs Critiques × 2) + (Délai Moy. × 0,5)
```

- Les ruptures sont pondérées le plus fortement (×3) car elles causent des arrêts de production
- Les SKUs critiques sont pondérés ×2 (sur le point de causer des arrêts)
- Les longs délais ajoutent un risque de base (×0,5)

**Pourquoi c'est important :** Les problèmes fournisseurs se traduisent directement en coûts d'approvisionnement d'urgence, frais de transport express et pertes de production. Le score de risque vous indique quelles relations fournisseurs auditer en priorité.

---

### 📌 VMI — Vendor Managed Inventory (Gestion Partagée des Stocks)

**En langage clair :** Au lieu que vous gériez quand et combien réapprovisionner auprès d'un fournisseur, c'est le fournisseur qui prend en charge le maintien de vos stocks à des niveaux convenus. Il surveille vos données de consommation et réapprovisionne automatiquement.

**Idéal pour :** Les articles à fort volume, demande stable, de classe A, avec un fournisseur de confiance.

**Avantage :** Réduit votre charge de travail de commande et réduit souvent les ruptures car le fournisseur a une meilleure visibilité sur son propre planning de production.

---

### 📌 MRO — Maintenance, Repair & Operations (Maintenance, Réparation et Exploitation)

**En langage clair :** Tous les consommables et pièces utilisés pour maintenir l'usine en fonctionnement qui ne font **pas** partie du produit final.

**Exemples :** Huile de machine, gants, produits de nettoyage, pièces de rechange, ampoules, outils.

Les articles MRO sont souvent de classe C dans l'analyse ABC, mais une rupture sur un article MRO critique (ex. un fusible spécifique) peut quand même arrêter une ligne de production entière. Une bonne gestion surveille les MRO critiques même si leur valeur totale est faible.

---

### 📌 Sous-Ensembles (Sub-Assemblies)

**En langage clair :** Des composants partiellement assemblés qui seront combinés plus tard pour fabriquer le produit final. Ils se situent entre les matières premières et les produits finis dans la hiérarchie de production.

**Exemple en fabrication :**
```
Acier + Boulons → Châssis Soudé (Sous-Ensemble) → Combiné avec Moteur + Boîtier → Machine Finale
```

L'optimisation des stocks de sous-ensembles est plus complexe car les ruptures peuvent se propager en cascade dans la chaîne de production.

## 📈 PARTIE 4 — Comprendre Chaque Graphique
---

### Graphique 1 — Santé des Stocks (Graphique en Anneau)
**Ce qu'il montre :** Quel pourcentage de vos SKUs relève de chaque catégorie de santé en ce moment.

**Ce qu'il faut chercher :** Une opération saine devrait avoir majoritairement du VERT (Sain). Trop de ROUGE (Rupture/Critique) signifie que l'approvisionnement est réactif, pas proactif. Trop de BLEU (Surstock) signifie que les achats ne sont pas synchronisés avec la demande réelle.

---

### Graphique 2 — Stock Actuel vs Jours de Couverture (Nuage de Points à Bulles)
**Ce qu'il montre :** Chaque SKU comme une bulle. Axe X = stock disponible. Axe Y = combien de jours ce stock durera. Taille de la bulle = coût unitaire (plus grande = article plus cher).

**Ce qu'il faut chercher :** Les grosses bulles (articles chers) qui sont EN DESSOUS de la ligne rouge des 7 jours représentent votre situation la plus dangereuse — articles à coût élevé sur le point d'être en rupture.

---

### Graphique 3 — QEC vs Stock Actuel vs Point de Réappro. (Barres Groupées)
**Ce qu'il montre :** Pour chaque SKU, une comparaison de trois chiffres clés côte à côte.

**Ce qu'il faut chercher :**
- Barre bleue (Stock Actuel) en dessous du diamant (Point de Réappro.) → commander MAINTENANT
- Barre bleue bien plus haute que la barre verte (QEC) → vous avez commandé trop la dernière fois

---

### Graphique 4 — Tendances de Demande Hebdomadaire (Courbes)
**Ce qu'il montre :** Comment la demande a évolué sur 52 semaines pour les 6 SKUs à plus forte valeur.

**Ce qu'il faut chercher :** Schémas saisonniers (vagues), pics soudains, ou tendances à la hausse/baisse. Ces schémas devraient guider la quantité de stock de sécurité et le calendrier de commande.

---

### Graphique 5 — Variabilité de la Demande / CV (Nuage de Points)
**Ce qu'il montre :** Chaque SKU tracé selon sa consommation journalière et son degré d'imprévisibilité. Couleur = catégorie, taille = coût de stockage.

**Ce qu'il faut chercher :** Les SKUs dans la zone en haut à droite (forte demande ET haute variabilité) sont vos articles à risque de rupture le plus élevé. Ils nécessitent plus de stock de sécurité.

---

### Graphique 6 — Risque Fournisseur vs Délai de Livraison (Bulle)
**Ce qu'il montre :** Chaque fournisseur tracé selon son délai moyen et son score de risque calculé. Taille de bulle = coût de stockage généré par leurs SKUs.

**Ce qu'il faut chercher :** Les fournisseurs en haut à droite (long délai + risque élevé) sont vos dépendances les plus dangereuses. Envisagez un double sourcing ou négociez de meilleures conditions.

---

### Graphique 7 — Répartition des Coûts de Stockage (Sunburst)
**Ce qu'il montre :** Un diagramme en cercles imbriqués montrant la décomposition des coûts de stockage annuels. Anneau extérieur = fournisseur. Anneau intérieur = catégorie.

**Ce qu'il faut chercher :** Quelle combinaison catégorie/fournisseur vous coûte le plus en stockage ? C'est là qu'une réduction des quantités commandées économiserait le plus d'argent.

---

### Graphique 8 — Pareto ABC (Barres + Courbe)
**Ce qu'il montre :** Les SKUs triés par valeur annuelle (barres), avec une courbe de pourcentage cumulé au-dessus.

**Ce qu'il faut chercher :** Le point où la courbe croise 70% marque la frontière entre les articles de classe A et B. Généralement, un petit nombre de SKUs représente la majorité de votre investissement en stocks.

---

### Graphique 9 — Carte de Chaleur des Stocks
**Ce qu'il montre :** Une grille colorée. Lignes = catégories (Matières Premières, Composants, etc.), Colonnes = statut de stock. Le chiffre dans chaque cellule = combien de SKUs sont dans cette combinaison.

**Ce qu'il faut chercher :** Quelle catégorie a le plus de ruptures ou d'articles critiques ? Cela vous indique où la planification des approvisionnements doit s'améliorer.

---

### Graphique 10 — Économies Potentielles sur le Surstock
**Ce qu'il montre :** Un graphique à barres horizontales de chaque SKU en surstock avec l'économie annuelle que vous réaliseriez en le réduisant à 70% de la capacité entrepôt.

**Ce qu'il faut chercher :** Le total indiqué dans le titre du graphique représente le montant total que vous pourriez économiser par an rien qu'en ne sur-commandant pas.

## 🎯 PARTIE 5 — Comment Fonctionne le Plan d'Action
---

La dernière section du notebook principal génère automatiquement une action recommandée pour chaque SKU qui nécessite une attention. Voici la logique de décision :

```
POUR CHAQUE SKU :

  SI Stock Actuel = 0
     → 🚨 COMMANDE URGENTE : Passer une commande de [QEC] unités immédiatement

  SINON SI Jours de Couverture < Délai de Livraison
     → ⚠️ RÉAPPRO. URGENT : Stock inférieur au tampon de délai

  SINON SI Jours de Couverture < 1,5 × Délai de Livraison
     → 📦 RÉAPPROVISIONNER BIENTÔT : Approche du point de réappro.

  SINON SI Stock Actuel > 85% de la Capacité Max
     → 💰 RÉDUIRE LE STOCK : Calculer les unités en excès et le coût annuel gaspillé

  EN PLUS :
  SI CV > 0,5 ET Classe ABC = A
     → 📈 HAUTE VOLATILITÉ : Augmenter le stock de sécurité ou mettre en place le VMI

  SI Délai de Livraison > 20 jours ET Statut est CRITIQUE ou BAS
     → 🏭 LONG DÉLAI : Envisager un double sourcing auprès d'un fournisseur de secours
```

Cette logique reproduit ce qu'un responsable supply chain expérimenté ferait manuellement — mais elle s'exécute sur l'ensemble des 40 SKUs en quelques secondes.

## 🔑 PARTIE 6 — Tableau de Référence des Formules Clés

| Formule | Équation |
|---|---|
| Jours de Couverture | Stock Actuel ÷ Demande Moy. Journalière |
| Point de Réapprovisionnement | (Demande Moy. Journalière × Délai de Livraison) + Stock de Sécurité |
| QEC (Quantité Écon. de Commande) | √( 2 × Demande Annuelle × Coût de Commande ÷ Coût de Stockage Unitaire ) |
| Coût de Stockage Unitaire | Coût Unitaire × Taux de Stockage (ex. 25%) |
| Coût de Stockage Annuel | Stock Actuel × Coût de Stockage Unitaire |
| Coefficient de Variation (CV) | Écart-Type de la Demande ÷ Demande Moyenne |
| Score de Risque Fournisseur | (Ruptures × 3) + (SKUs Critiques × 2) + (Délai Moy. × 0,5) |
| Valeur Annuelle (ABC) | Demande Annuelle × Coût Unitaire |
